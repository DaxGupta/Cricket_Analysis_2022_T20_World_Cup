<h1>T20i Cricket World Cup Data Pre Processing</h1>

```python
import pandas as pd
import json
```

<h2>Match Scorecard<br>
<i>From JSON to be Shown in DataFrame</i></h2>

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

<h3>Changing Column Name From ScoreCard to Match_id</h3>

```python
df_match.rename({'scorecard': 'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
df_match.to_csv('t20_csv_file/match_summary.csv',index=False)
```

<h2>Batting Summary</h2>

```python
with open('t20_json_files/t20_wc_batting_summary.json') as m:
    data = json.load(m)

all_records =[]

for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting=pd.DataFrame(all_records)
df_batting.tail(17)
```

<h3>Adding Another Column Namely Out/Not_Out</h3>

```python
df_batting["out/not_out"]= df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out" )
```

```python
df_batting.tail(17)
```

<h3>Cleaning The Data by altering the column namely Batsman Table</h3>

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting.tail(17)
```

```python
match_ids_dict= {}

for index,row in df_match.iterrows():
    key1 = row['team1']+' Vs '+row['team2']
    key2 = row['team2']+' Vs '+row['team1']

    match_ids_dict[key1]= row["match_id"]
    match_ids_dict[key2]= row["match_id"]

match_ids_dict
```

```python
df_batting["mach_id"]= df_batting["match"].map(match_ids_dict)
df_batting.tail(17)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.tail(17)
```

```python
df_batting['dismissal']= df_batting['dismissal'].apply(lambda x: x.replace('â€',''))
df_batting.tail(17)
```

```python
df_batting.shape
```

```python
df_batting.to_csv('t20_csv_file/batting_summary.csv',index=False)
```

<h2>Bowling Summary</h2>

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec['bowlingSummary'])

all_records[:10]
```

```python
df_bowling=  pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id']=  df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_file/bowling_summary.csv', index= False)
```

<h2>Players Information</h2>

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players=  pd.DataFrame(data)
print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv('t20_csv_file/players_info_no_image.csv',index=False)
```
